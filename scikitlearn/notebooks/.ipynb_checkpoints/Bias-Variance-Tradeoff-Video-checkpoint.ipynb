# In one step

```python
ndata = 1000
nsamps = 100

xdata = np.linspace(-1, 1, ndata)
ydata = np.sin(np.pi*xdata)  # Training model
data = np.asarray(list(zip(xdata, ydata)))


ranData_1 = []
ranData_2 = []
midpoints = []

for i in range(0, nsamps):
    x1,y1 = data[ data[:,1] == random.choice(ydata) ][0]
    x2,y2 = data[ data[:,1] == random.choice(ydata) ][0]
    ranData_1.append( [x1,y1] )
    ranData_2.append( [x2,y2] )
    midpoints.append( np.average([y1,y2]) )
```

```python
m = np.average(midpoints) # very close to 0.
#qtls = np.percentile(midpoints, q = [50-(68.27/2), 50+(68.27/2)]) # 1sigma quantiles, q = [15.865, 84.135] 
qtls = np.percentile(midpoints, q = [50-(68/2), 50+(68/2)]) # 1sigma quantiles, q = [15.865, 84.135] 

print(m)
print(qtls)
```

```python
abs(np.diff([qtls[0], m])[0]), abs(np.diff([qtls[1], m])[0])
```

```python
stdev = np.std(midpoints) # standard deviation
print(stdev)
```

```python
# variance = std^2
var = stdev**2
print(var)
```

```python
# width of the grey region.
np.diff([qtls])  # or qtls[1]-qtls[0]
```

In his video he states the variance is the width of the grey region. This doesn't make sense if the horizontal lines and shaded region are 1-sigma away from the average. 

```python
print(m-np.std(midpoints))
print(m+np.std(midpoints))
```

```python
plt.figure(figsize=(11,7))
plt.plot(xdata, ydata, label='target function')
plt.hlines(m, -1, 1, color='lime', linestyle='-', alpha=1, label='average of constant models')
plt.fill_between(x=[-1,1], y1=qtls[0], y2=qtls[1], color='grey', alpha=0.3 )
plt.hlines(qtls, -1, 1, color='grey', linestyle='-', alpha=0.75, label='quantiles (1-sigma)')
plt.hlines([m-stdev, m+stdev], -1, 1, color='black', linestyle='-.', alpha=0.75, label='standard deviation')
plt.legend(loc='best', numpoints=1)
plt.show()
```

Width of the grey area gives you the vairance. 

# Fitting the linear model

```python
def fit_line(xvalues, yvalues):
    '''
    fit_line(xvalues, yvalues)
    
    Ordinary Least Squares Linar Model fitting function. 
    
    PARAMETERS:
    ----------
    xvalues: list of x-axis values for data points. [x1, x2] in this case.
    yvalues: list of y-axis values for data points. [y1, y2] in this case.
    
    both must be in list format before passing to this function. 
    
    RETURN:
    ----------
    Returns a matrix of linear modeling coefficients. First is slope, 
    second is the y-intercept. 
    
    '''
    n = len(xvalues)
    X = np.matrix([np.asarray(xvalues), np.ones(n)]).T
    y = np.asarray(yvalues)
    return ((X.T * X)**-1) * X.T * np.vstack(y)
```

```python
ndata = 1000
nsamps = 100

xdata = np.linspace(-1, 1, ndata)
ydata = np.sin(np.pi*xdata)  # Training model

data = np.asarray(list(zip(xdata, ydata)))


ranData_1 = []
ranData_2 = []
midpoints = []

all_coeffs = []
for i in range(0, nsamps):
    x1,y1 = data[ data[:,1] == random.choice(ydata) ][0]
    x2,y2 = data[ data[:,1] == random.choice(ydata) ][0]
    ranData_1.append( [x1,y1] )
    ranData_2.append( [x2,y2] )
    # FIT A LINE TO THE TWO DATA POINTS. RETURN COEFFICIENTS. 
    coeffs = fit_line([x1,x2], [y1,y2])  # returns in matrix form.
    all_coeffs.append( np.asarray(coeffs.flatten())[0] )

all_coeffs = np.vstack(all_coeffs) 
# all_coeffs[:,0]  is all the slopes
# all_coeffs[:,1]  is all the y-intercepts
```

```python
xarr = np.linspace(-1.1, 1.1, 10)

plt.clf()
plt.figure(figsize=(9,7))
for m,b in all_coeffs:
    plt.plot(xarr, m*xarr+b, 'g-', alpha=0.25)
plt.plot(xdata, ydata, color='blue')
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.show()
```

```python
xarr = np.linspace(-1.1, 1.1, 10)
m,b = np.mean(all_coeffs[:,0]), np.mean(all_coeffs[:,1])

plt.clf()
plt.figure(figsize=(9,7))
plt.plot(xarr, m*xarr+b, 'g-', alpha=0.75)
plt.plot(xdata, ydata, color='blue')
plt.hlines(0, -1, 1, color='grey', linestyle='-', alpha=0.4)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.show()
```
